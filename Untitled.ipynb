```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

%matplotlib inline 

mini_batch_size = 50
# np.info('randn')
data_x = np.random.uniform(-3.0,3.0,100)
num_epcho = 100 // mini_batch_size

print(data_x)
data_y = 2. * data_x + 0.3 + np.random.randn(100,)

g1 = tf.Graph()
with g1.as_default():
    with tf.variable_scope('LG') as LG_SCOPE:
        w = tf.Variable(0.,name = 'w')
        b = tf.Variable(0.01,name = 'b')
        x = tf.placeholder(shape=(None,),dtype = tf.float32)
        y = tf.placeholder(shape=(None,),dtype = tf.float32)
    
#     y_hat = tf.add(tf.multiply(w,x),b)
    y_hat = tf.multiply(w,x) + b
    loss = tf.reduce_mean((y_hat - y)**2)
    opt = tf.train.AdamOptimizer(0.001).minimize(loss)
    
with tf.Session(graph = g1)  as sess:
    init_op = tf.global_variables_initializer()
    sess.run(init_op)
    
    for i in range(1000):
        batch_idx = i % num_epcho
#         _,loss_,w_,b_= sess.run([opt,loss,w,b],feed_dict ={x:data_x[mini_batch_size*batch_idx:mini_batch_size*batch_idx+mini_batch_size],
#                                                            y:data_y[mini_batch_size*batch_idx:mini_batch_size*batch_idx+mini_batch_size]})
        _,loss_,w_,b_= sess.run([opt,loss,w,b],feed_dict ={x:data_x,y:data_y})
        if (i + 1) % 10 == 0:
            print('loss of {0}th iteration is: {1}'.format(i,loss_))

plt.scatter(data_x,data_y)
x_baseline = np.arange(-3.0,3.0,0.01)
y_baseline = w_*x_baseline + b_
print('w is :{0} and b is {1}'.format(w_,b_))
plt.plot(x_baseline,y_baseline)
plt.show()
```
